```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

## LINE PLOT

```python
x=[1,2,3,4]
y=[10,20,15,26]
plt.plot(x,y)

plt.title("data science")

plt.show()
```

```python
x=['sun','mon','tues','wed']        # x-axis

y=[10,50,20,40]       #y-axis

plt.plot(x,y, color="black",linestyle="--",linewidth=5,marker="o")

plt.title("Bakery sales this week!")

plt.xlabel("day of this week")

plt.ylabel("sale per day")

plt.show()
```

```python
month=[1,2,3,4,5]

sales=[100,600,300,800,500]

plt.plot(month,sales, color="black",marker="o",linestyle="--",linewidth="4",label="2025 saled data")

plt.xlabel("months")

plt.ylabel("sales per month")

plt.title("monthly sales data report")

plt.legend()

plt.show()
```

```python
college=[1,2,3,4]
student=[14,12,13,11]

plt.plot(college,student,color="black",marker="d",linestyle="--",linewidth=3,label="2025 class information")

plt.title("college information")

plt.xlabel("classes")

plt.ylabel("class per student")

plt.legend(loc="upper left",fontsize=10)

plt.grid(color="blue",linestyle="--",linewidth=1)

plt.xlim(1,4)

plt.ylim(11,15)

plt.xticks([1,2,3,4],['m1','m2','m3','m4'])

plt.yticks([11,12,13,14,15],['a1','a2','a3','a4','a5'])

plt.show()
```

```python
product=["A","B","C","D","E"]

sales=[100,500,300,700,600]

plt.plot(product,sales,color="blue",marker="o",linestyle="--",linewidth=4,label="2025 sales")

plt.title("product sale in 2025")

plt.ylabel("sales")

plt.xlabel("product")

plt.grid(color="black",linewidth=2)

plt.xlim("A","E")

plt.ylim(100,800)

plt.xticks(["A","B","C","D","E"],["iphone","vivo","nike","realme","oppo"])

plt.legend()

plt.show()
```

```python
company=["apple","microsoft","amazon","tesla","meta platforms"]

employee=[300,966,84,892,895]

plt.plot(company,employee,color="black",linestyle="--",linewidth=3,marker="o",label="data 2024")

plt.legend(fontsize=10)

plt.title("company and employee")

plt.xlabel("company")

plt.ylabel("employee")

plt.grid(color="red",linewidth=1)

plt.show()
```

## BAR CHART

```python
product=["iphone","vivo","nike","realme","oppo"]

sales=[1000,600,300,900,800]

plt.bar(product,sales,color="blue",label="sales 2025")

plt.xlabel("product")

plt.ylabel("sales")

plt.title("product sale comparison")

plt.legend(fontsize=15)

plt.show()
```

```python
company=["apple","microsoft","amazon","tesla","meta platforms"]

employee=[300,966,500,700,895]

plt.barh(company,employee,color="green",label="data 2024")

plt.legend(fontsize=10)

plt.title("company and employee")

plt.xlabel("company")

plt.ylabel("employee")

plt.show()
```

## PIE CHART

```python
regions=["north","south","east","west"]

revenue=[3000,1500,1000,2000]

plt.pie(revenue,labels=regions,autopct="%1.1f%%",colors=["gold","skyblue","lightgreen","blue"])

plt.title("revenue contribution by region")

plt.show()
```

## SCATTER PLOT

```python
hours_student=[1,2,3,4,5,6,7,8]

exam_scores=[50,55,60,65,70,75,80,85]

plt.scatter(hours_student,exam_scores,color="red",label="student data",marker="p")

plt.xlabel("Hours student")

plt.ylabel("Exam Scores")

plt.title("Relationship Between Time and Exam Score")

plt.grid(color="black")

plt.legend(fontsize=10)

plt.show()
```

```python
hours_student=[1,2,3,4,5,6,7,8]

exam_scores=[50,55,60,65,70,75,80,85]

plt.scatter(hours_student,exam_scores,color="red",label="student data",marker="*")

plt.xlabel("Hours student")

plt.ylabel("Exam Scores")

plt.title("Relationship Between Time and Exam Score")

plt.grid(True)

plt.legend(fontsize=10)

plt.show()
```

```python
plt.scatter([1,2,3,4,5],[10,20,30,40,50],color="green",label="Scores A")

plt.scatter([1,2,3,4,5],[30,40,50,60,70],color="blue",label="Scores B")

plt.xlabel("Hours student")

plt.title("comparison of two student")

plt.ylabel("Exam Score")

plt.grid(True)

plt.legend()

plt.show()
```

## SUBPLOTS AND LAYOUT ADJUSTMENTS

```python
x=[1,2,3,4]

y=[10,20,15,25]

plt.subplot(1,2,1)    # 1row, 2column, 1subplot

plt.plot(x,y)

plt.title("Line Chart")

plt.subplot(1,2,2)     #1row, @column ,1sublpot

plt.bar(x,y)

plt.title("Bar chart")

plt.tight_layout()

plt.show()
```

```python
#fig.ax=plt.subplot(no row,no column,figsize=width,height)

fig,ax=plt.subplots(1,2,figsize=(10,5))

x=[1,2,3,4]

y=[10,20,15,25]

ax[0].plot(x,y,color="green")

ax[0].set_title("Line plot")

ax[1].bar(x,y,color="red")

ax[1].set_title("Bar chart")

plt.suptitle("Comparison of Line and Bar Chart")

plt.tight_layout()

plt.show()
```

## SAVING FIGURES

```python
x=[1,2,3,4]

y=[10,20,15,25]

#create plot

plt.plot(x,y,color="black",marker="o",linestyle="--",linewidth=2)

plt.title("Simple Line Plot")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.grid()

plt.savefig("line_plot.png",dpi=300,bbox_inches='tight')

plt.show()
```
